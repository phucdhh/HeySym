## Demo: T√≠ch h·ª£p v·ªõi SymPy

````python
from sympy import symbols, diff, integrate, solve, simplify
from sympy import init_printing
init_printing()

# ƒê·ªãnh nghƒ©a bi·∫øn
x = symbols('x')

# B√†i to√°n: T√¨m c·ª±c tr·ªã c·ªßa h√†m s·ªë y = x¬≥ - 3x¬≤ + 2
y = x**3 - 3*x**2 + 2

print("H√†m s·ªë:", y)

# T√≠nh ƒë·∫°o h√†m
y_prime = diff(y, x)
print("ƒê·∫°o h√†m:", y_prime)

# T√¨m ƒëi·ªÉm c·ª±c tr·ªã (y' = 0)
critical_points = solve(y_prime, x)
print("C√°c ƒëi·ªÉm t·ªõi h·∫°n:", critical_points)

print("\n" + "="*60)
print("Gi·∫£i th√≠ch t·ª´ AI:")
print("="*60)

# H·ªèi AI gi·∫£i th√≠ch - output t·ª± ƒë·ªông streaming v√† formatted
explanation = ai.ask(f"""
T√¥i ƒëang t√¨m c·ª±c tr·ªã c·ªßa h√†m s·ªë y = x¬≥ - 3x¬≤ + 2.
ƒê·∫°o h√†m: y' = {y_prime}
C√°c ƒëi·ªÉm t·ªõi h·∫°n: {critical_points}

H√£y gi·∫£i th√≠ch b·∫±ng ti·∫øng Vi·ªát:
1. C√°c ƒëi·ªÉm n√†y l√† c·ª±c ƒë·∫°i hay c·ª±c ti·ªÉu?
2. Gi√° tr·ªã h√†m s·ªë t·∫°i c√°c ƒëi·ªÉm ƒë√≥ l√† bao nhi√™u?
3. √ù nghƒ©a h√¨nh h·ªçc?
""")
````
